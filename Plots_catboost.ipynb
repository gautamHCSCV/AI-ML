```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

# FBL

```python
# RBF VS CATBOOST ON FBL
rbf_time = [6.89,7.89,7.91,7.106,7.43,6.86,7.95,7.37,7.81]
catf_time = [4.55,4.73,5.01,4.61,4.68,6.22,4.99,4.50,4.62]
rf_time = [11.49,12.87,13.97,11.88,8.5,14.74,18.16,18.15,9.73]
catr_time = [4.61,4.65,4.66,4.6,4.51,4.51,4.53,4.38,4.43]

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.set_figheight(4.5)
fig.set_figwidth(6)
ax1.grid(linestyle='--',linewidth=0.8)
ax1.plot(range(1,10),rbf_time,label='RBF-FBL')
ax1.plot(range(1,10),catf_time,label='CAT-FBL')
ax1.legend()
#ax1.set_xlabel('Object',fontsize=16)
ax1.set_ylabel('Time (in sec)',fontsize=16)
ax1.set_xlim([1,9])
ax1.set_ylim([2,9])

ax2.grid(linestyle='--',linewidth=0.8)
ax2.plot(range(1,10),rf_time,label='RF-RS')
ax2.plot(range(1,10),catr_time,label='CAT-RS')
ax2.legend()
ax2.set_xlabel('Object',fontsize=16)
ax2.set_ylabel('Time (in sec)',fontsize=16)
ax2.set_xlim([1,9])
ax2.set_ylim([3,20])
#plt.savefig('ttime.pdf')
plt.show()
```

```python
rbf_errors = [0.0753,0.08,0.0844,0.0840,0.0956,0.0977,0.102,0.097,0.11]
cat_errors = [0.058,0.077,0.0686,0.0761,0.0713,0.0781,0.0925,0.0842,0.0913]
plt.plot(range(1,10),rbf_errors,label='RBF')
plt.plot(range(1,10),cat_errors,label='Catboost')
plt.legend()
plt.title('Root Mean Squared Errors Comparison')
plt.xlabel('Object')
plt.ylabel('RMSE')
plt.xlim([1,9])
plt.ylim([0.02,0.14])
plt.show()
```
