```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
np.arange(4)
```

# FBL

```python
# RBF VS CATBOOST ON FBL
rbf_time = [6.89,7.89,7.91,7.106,7.43,6.86,7.95,7.37,7.81]
catf_time = [4.55,4.73,5.01,4.61,4.68,6.22,4.99,4.50,4.62]
rf_time = [11.49,12.87,13.97,11.88,12.5,14.74,16.16,16.15,16.73]
catr_time = [4.61,4.65,4.66,4.6,4.51,4.51,4.53,4.38,4.43]

X = np.arange(1,10)
figure(figsize=(8, 6), dpi=80)

plt.grid(linestyle='--',linewidth=0.8)
plt.bar(X-0.25, rbf_time, label = 'RBF',width=0.25,color='blue')
plt.bar(X+0.00, rf_time,label = 'Random Forest',width=0.25,color='green')
plt.bar(X+0.25, catr_time, label = 'Catboost',width=0.25,color='orange')

plt.legend(loc = 'upper left')
plt.xlabel('Object',fontsize=16)
plt.ylabel('Time (in sec)',fontsize=16)
plt.xlim([0.5,9.5])
plt.ylim([3,20])
plt.savefig('ttime_bar.pdf',format='pdf')
plt.show()
```

```python
rbf_errors = [0.0753,0.08,0.0844,0.0840,0.0956,0.0977,0.102,0.097,0.11]
cat_errors = [0.058,0.077,0.0686,0.0761,0.0713,0.0781,0.0925,0.0842,0.0913]
plt.plot(range(1,10),rbf_errors,label='RBF')
plt.plot(range(1,10),cat_errors,label='Catboost')
plt.legend()
plt.title('Root Mean Squared Errors Comparison')
plt.xlabel('Object')
plt.ylabel('RMSE')
plt.xlim([1,9])
plt.ylim([0.02,0.14])
plt.show()
```
